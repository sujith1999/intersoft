```python
normals.cumsum().plot(grid=True)
```

```python
variable= pd.DataFrame({"normal":np.random.normal(size=10),
                       "gamma":np.random.gamma(1,size=10),
                       "poisson": np.random.poisson(size=10) })
```

```python
print(variable)
```

```python
variable.cumsum(0).plot()
```

```python
variable.cumsum(1).plot(subplots=True)
```

```python
variable.cumsum(0).plot(secondary_y="gamma")
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(4, 12))
for i,var in enumerate(['normal','gamma','poisson']):
    variable[var].cumsum(0).plot(ax=axes[i], title=var)
axes[0].set_ylabel('cumulative sum')
```

```python
fig = plt.figure()
ax = plt.axes()

s = np.linspace(0, 20, 100)
print(s)
ax.plot(s, np.sin(s));
```

```python
plt.plot(s, np.sin(s - 0), linestyle='-')        
plt.plot(s, np.sin(s - 1), linestyle=':')           
plt.plot(s, np.sin(s - 2), linestyle='-.')  
```

```python
rng = np.random.RandomState(7)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5,
            cmap='Reds')
plt.colorbar()
```
